### Jawaban Soal 3

Pada case soal no 3 saya memilih menggunakan Bar Chart karena pada soal ini diminta untuk menunjukkan visualisasi data yang menunjukkan popularitas relatif. Dengan menggunakan Bar Chart bisa dengan cepat untuk membandingkan data dari seluruh kategori yang ada. Kita akan mudah untuk membandingkan nilai antar beberapa kategori untuk menunjukkan nilai terkecil dan terbesar dari suatu data. 

Setelah dilakukan visualisasi dapat dilihat bahwa makanan penutup yang harus disingkirkan dari daftar makanan adalah Kue Wajik, Pastel, dan Pudding Vanila.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datapenjualan_makananpenutup =  {
    'Donat': 14,  
    'Pastel': 5,  
    'Kue Coklat': 12,  
    'Ice Cream': 19,  
    'Puding Vanila': 6,  
    'Brownies': 8,  
    'Puding Strawberi': 12,  
    'Puding Coklat': 9,  
    'Ice Cream Nutela': 10, 
    'Kue Coklat-Keju': 17,  
    'Kue Wajik': 2,  
    'Kue Sus': 9,  
    'Mochi': 13, 
}

namaMakanan = list(datapenjualan_makananpenutup.keys())
jumlahPenjualan = list(datapenjualan_makananpenutup.values())

# data frame
df = pd.DataFrame({'Makanan': namaMakanan, 'Jumlah': jumlahPenjualan,})
df.sort_values(by='Jumlah', inplace = True, ascending = False)

# len ubah berdasar df
kor_x = np.arange(len(df)) 
warna = ['#0000FF' for _ in range(len(df))]
warna [-1] = '#FF0000'
warna [-2] = '#FF0000'
warna [-3] = '#FF0000'

plt.figure(figsize=(15,7))

plt.bar(kor_x, df['Jumlah'], tick_label=df['Makanan'], color=warna)
plt.xticks(rotation=90) 
plt.ylabel('Makanan')
plt.xlabel('Penjualan')

plt.title('Data Penjualan Makanan Penutup')
plt.show()
```
